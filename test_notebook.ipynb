```python
import random
import json
```

```python
SUCCESS_CODE = 0
WARNING_CODE = 1
ERROR_CODE = 2

OUTPUT_STATUS = {
  SUCCESS_CODE: "SUCCESS",
  WARNING_CODE: "WARNING",
  ERROR_CODE: "ERROR"}

OUTPUT_MESSAGES = {
  SUCCESS_CODE: "Notebook ended with SUCCESS",
  WARNING_CODE: "Notebook ended with WARNING",
  ERROR_CODE: "Notebook ended with ERROR"}  

class WarningException(Exception):
    pass   


class ErrorException(Exception):
    pass 
```

```python
# generarate a random number between 0 and 2
def generate_random_number():
  return random.randint(0,2)

# generate a random message based on the random number
def generate_random_message(number):
  return OUTPUT_MESSAGES[number]

# generate a random status based on the random number
def generate_random_status(number):
  return OUTPUT_STATUS[number]
```

```python
try:
    generated_output = generate_random_number()
    print (OUTPUT_MESSAGES[generated_output])
    if generated_output == SUCCESS_CODE:
        dbutils.notebook.exit(json.dumps( OUTPUT_STATUS[generated_output] ))
    elif generated_output == WARNING_CODE:
        raise WarningException
    else:
        raise ErrorException
except WarningException as e1:
    dbutils.jobs.taskValues.set(key="error_code", value=OUTPUT_STATUS[generated_output])
    dbutils.jobs.taskValues.set(key="error_message", value=OUTPUT_MESSAGES[generated_output])
    raise
except ErrorException as e2:
    dbutils.jobs.taskValues.set(key="error_code", value=OUTPUT_STATUS[generated_output])
    dbutils.jobs.taskValues.set(key="error_message", value=OUTPUT_MESSAGES[generated_output])
    raise
```
